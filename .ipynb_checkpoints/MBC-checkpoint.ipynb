```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.mixture import GaussianMixture
from sklearn.metrics import confusion_matrix
```

#### Description of the dataset

The dataset is available in R as one of built-in datasets. <br>
<br>

**Name**: Swiss banknotes data <br>
**Description**: The data set contains six measurements made on 100 genuine and 100 counterfeit old-Swiss 1000-franc bank notes. <br><br>
**Format** <br>
A data frame with the following variables: <br>
<br>
*Status* <br>
the status of the banknote: genuine or counterfeit<br>
<br>
*Length*<br>
Length of bill (mm)<br>
<br>
*Left*<br>
Width of left edge (mm)<br>
<br>
*Right*<br>
Width of right edge (mm)<br>
<br>
*Bottom*<br>
Bottom margin width (mm)<br>
<br>
*Top*<br>
Top margin width (mm)<br>
<br>
*Diagonal*<br>
Length of diagonal (mm)<br>
<br>
**Source**<br>
Flury, B. and Riedwyl, H. (1988). Multivariate Statistics: A practical approach. London: Chapman & Hall, Tables 1.1 and 1.2, pp. 5-8.

Scatterplot of the data. Color assigned to the Status attribute. Points jittered to avoid overplotting.

#### Encoding categorical variable Status

```python
Status_encoded = []
for i in range(len(banknote)):
    if banknote.iloc[i]["Status"].strip() == "genuine":
        Status_encoded.append(1)
    else:
        Status_encoded.append(0)

# Create a new attribute in the dataframe for the encoded values.
banknote["Status_encoded"] = Status_encoded
```

## Unsupervised Learning

### Removing the label from the dataframe. Preparing the dataframe for clustering

```python
df = banknote.drop(columns=["Status", "Status_encoded"])
df.head()
```

# Univariate model-based clustering

```python
diagonal = df["Diagonal"].to_numpy().reshape(-1,1)
```

```python
# Model's covariance type: all components share the same general covariance matrix
gmm_uni = GaussianMixture(n_components=2, covariance_type="tied")
# fit the model on the training data
gmm_uni.fit(diagonal)
```

```python
# AIC and BIC to evaluate how well the model fits the data.
print("Aikake information criterion (AIC) score: %f\nBayesian information criterion (BIC) score: %f" % (gmm_uni.aic(diagonal), gmm_uni.bic(diagonal)))
```

```python
labels_uni = gmm_uni.predict(diagonal)
# switch the encoding
labels_uni = abs(labels_uni - 1)
```

```python
cm = confusion_matrix(banknote.Status_encoded.to_numpy(), labels_uni)
sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('True label')
plt.ylabel('Predicted label')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

print(accuracy_score(banknote.Status_encoded, labels_uni))
print(f1_score(banknote.Status_encoded, labels_uni))
```

# Multivariate model-based clustering

```python
# Model-based clustering with 2 clusters on Diagonal feature
```

```python
# Fitting the model
```

```python
# Model's covariance type: all components share the same general covariance matrix
gmm_E_2 = GaussianMixture(n_components=2, covariance_type="tied")
# fit the model on the training data
gmm_E_2.fit(df)
```

```python
# AIC and BIC to evaluate how well the model fits the data.
print("Aikake information criterion (AIC) score: %f\nBayesian information criterion (BIC) score: %f" % (gmm_E_2.aic(df), gmm_E_2.bic(df)))
```

```python
n_c = np.arange(1, 21)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(df)
          for n in n_c]

plt.plot(n_c, [m.bic(df) for m in models], label='BIC')
plt.plot(n_c, [m.aic(df) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
```

```python
labels = gmm_E_2.predict(df)
```

```python
# switch the encoding
labels = abs(labels-1)
```

```python
cm = confusion_matrix(banknote.Status_encoded.to_numpy(), labels)
sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('True label')
plt.ylabel('Predicted label')
```

```python
# The clustering done by the model is similar to the "true clusters". 
# the accuracy is very high

tp = cm[0][0]
fp = cm[1][0]
fn = cm[1][0]
tn = cm[1][1]

accuracy = (tp + tn) / (tp + tn + fp + fn)
print("Accuracy of the model on the dataset: %0.5f" % accuracy)
```

```python
precision = tp / (tp + fp)
sensitivity = tp / (tp + fn) # recall
specificity = tn / (tn + fp)

balanced_accuracy = (sensitivity + specificity) / 2

beta = 1
f_score = ((1 + beta ** 2) * (precision * sensitivity)) / ((beta ** 2) * precision + sensitivity)


print("Balanced accuracy of the model: %0.5f" % balanced_accuracy)
print("F1 score: %0.5f" % f_score)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

print(accuracy_score(banknote.Status_encoded, labels))
print(f1_score(banknote.Status_encoded, labels))
```

```python
plt.rcParams['figure.figsize'] = [15, 10]

figure, axis = plt.subplots(2)
axis[0].scatter(jittered_x, jittered_y, 
            c=labels, 
            cmap='viridis',
            s=100,
            edgecolors='b')

axis[0].set_title("Predicted")
  
# For Cosine Function
axis[1].scatter(jittered_x, jittered_y, 
            c=banknote.Status_encoded, 
            cmap='viridis',
            s=100,
            edgecolors='b')

axis[1].set_title("Initial clustering")


plt.show()
```
